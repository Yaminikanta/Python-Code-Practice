```python
# leetcode problem no 1295

def findNumbers(nums):
        count = 0
        for num in nums:
            if len(str(num)) % 2 == 0:
                count += 1
        return count
findNumbers([12,345,2,6,7896])
```

```python
# leetcode problem no 66

def sol1(digits):
    d1 = ""
    for i in digits:
        d1 += str(i)
        num = int(d1) + 1
        result = list(map(int, str(num)))
    return result
sol1([4,3,2,1])
```
