```python
# Day 1: Reverse String
# Question: Reverse a given string without using built-in functions.
# Source: GeeksforGeeks

class Solution:
     def reverseString(self, s: str) -> str:
        rev=""
        for i in s:
            rev=i+rev
        return rev
        
# Example
s = "hello"
obj = Solution()
print(obj.reverseString(s))
```

```python
# other method using inbuild function 

# Day 1: Reverse String (Using Slicing)
# Question: Reverse a given string without using built-in functions.
# Source: GeeksforGeeks

class Solution:
    def reverseString(self, s: str) -> str:
        return s[::-1]   # slicing method to reverse

# Example
s = "hello"
obj = Solution()
print(obj.reverseString(s))
```
