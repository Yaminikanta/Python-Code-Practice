##### Q.1 - Reverse a given string without using built-in functions.

```python
def reverseString(s):
    rev=""
    for i in s:
         rev=i+rev
    return rev  
     
print(reverseString("String"))

str = "apple"
print(reverseString(str))
```

```python
# other method using Using Slicing(build-in function)
def reverseString(b):
    return b[::-1]        # slicing method to reverse

print(reverseString("banana"))
```

##### Q.2 - Convert the string to lower case

```python
def toLower (s):
    s1=""
    for i in s:
        if 65 <= ord(i) <= 90:                          # By using ASCII Value
            s1 += chr(ord(i)+32)
        else:
            s1+=i
                
    return s1

print(toLower("GHKKKgggg"))
```

```python
# using inbuild function

def tolower(s):
    return s.lower()

print(tolower("QWERtyuioP"))
```

##### Q.3 - W.A.F to convert a string into uppercase without using upper().

```python
def toUpper(x):
    s1=""
    for i in x:
        if 97 <= ord(i) <= 122:                          # By using ASCII Value
            s1 += chr(ord(i)-32)
        else:
            s1+=i
                
    return s1

print(toUpper("GHKKKgggg"))
```

```python
# using inbuild function

def toUpper(s):
    return s.upper()

print(toUpper("QWERtyuioP"))
```

##### Q.3 - W.A.F to swap the case of characters in a string without using swapcase()

```python
def manual_swapcase(s):
    result = ""
    for char in s:
        if 'a' <= char <= 'z':                         # lowercase
            result += chr(ord(char) - 32)
        elif 'A' <= char <= 'Z':                       # uppercase
            result += chr(ord(char) + 32)
        else:
            result += char                             # non-alphabet stays same
    return result

text = "HeLLo WoRLd 123!"
print(manual_swapcase(text))
```

##### 4. W.A.F to convert the first letter of each word in a string to uppercase and the last lowercase without using title().

```python
def manual_title(s):
    words = s.split()
    result_words = []

    for w in words:
        if len(w) > 0:
            ch = w[0]                                   # Convert first letter to uppercase using ASCII
            if 'a' <= ch <= 'z':
                first = chr(ord(ch) - 32)
            else:
                first = ch

            
            last = ""                                   # Convert remaining letters to lowercase using ASCII
            for c in w[1:]:
                if 'A' <= c <= 'Z':
                    last += chr(ord(c) + 32)
                else:
                    last += c

            result_words.append(first + last)
        else:
            result_words.append(w)

    return " ".join(result_words)

print(manual_title('hEllO gooD NighT'))
```

```python
# other solution

def manual_title(s):
    words = s.split()                                # split by spaces
    result_words = []

    for w in words:
        if len(w) > 0:
            first = w[0].upper()                     # first letter → uppercase
            last = w[1:].lower()                     # remaining → lowercase
            result_words.append(first + last)
        else:
            result_words.append(w)

    return " ".join(result_words)

print(manual_title('hEllO gooD MorNing'))
```

##### 5. W.A.F to capitalize the first character of a string and convert the rest into lowercase without using capitalize().

```python
def manual_capitalize(s):
        
    first = s[0]
    if 'a' <= first <= 'z':
        first = chr(ord(first) - 32)

    last = ""
    for i in s[1:]:
        if 'A' <= i <= 'Z':
            last += chr(ord(i) + 32)
        else:
            last += i


    return first + last
            

print(manual_capitalize('heLLO GoodMorninG'))
```

##### 6. W.A.F to split a string into words without using split().

```python
def manual_split(s):
    words = []
    current_word = ""

    for ch in s:
        if ch != " ":
            current_word += ch
        else:
            if current_word != "":
                words.append(current_word)
                current_word = ""

    if current_word != "":
        words.append(current_word)

    return words

print(manual_split("hello world this is python"))
print(manual_split("hello world this is python ."))
```
